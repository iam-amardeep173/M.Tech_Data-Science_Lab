```python
# Prepare python program to take the vertices and edges from the user for any one of the following graphs and do as per the instructions:    
# a) Prepare the incidence matrix and incidence list of the given graph.
# b) A list of edges incident to given vertex.
# c) Count and List of vertices connected with incidence of > 2.
# d) Return the permutation graph with incidence of 2.

import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
from itertools import permutations

def create_graph():
    # Take the number of vertices from the user
    num_vertices = int(input("Enter the number of vertices: "))

    # Create an empty graph
    G = nx.Graph()

    # Add vertices to the graph
    for i in range(1, num_vertices + 1):
        G.add_node(i)

    # Take the number of edges from the user
    num_edges = int(input("Enter the number of edges: "))

    # Add edges to the graph
    print("Enter edges (e.g., 1 2 for an edge between vertex 1 and vertex 2):")
    for _ in range(num_edges):
        edge = input().split()
        G.add_edge(int(edge[0]), int(edge[1]))

    return G

def live_show(graph):
    # Draw the initial graph
    pos = nx.spring_layout(graph)
    nx.draw(graph, pos, with_labels=True, font_weight='bold')
    plt.show(block=False)

    # Keep the plot window open
    plt.pause(0.001)

    # Wait for the user to close the plot window
    input("Press Enter to close the live show.")

def incidence_matrix(graph):
    num_vertices = len(graph.nodes)
    num_edges = len(graph.edges)

    # Initialize the incidence matrix with zeros
    matrix = np.zeros((num_vertices, num_edges), dtype=int)

    # Fill in the incidence matrix
    for i, edge in enumerate(graph.edges):
        matrix[edge[0] - 1, i] = 1
        matrix[edge[1] - 1, i] = -1

    return matrix

def incidence_list(graph):
    incidence_dict = {vertex: [] for vertex in graph.nodes}

    # Fill in the incidence list
    for i, edge in enumerate(graph.edges):
        incidence_dict[edge[0]].append(i + 1)
        incidence_dict[edge[1]].append(i + 1)

    return incidence_dict

def edges_incident_to_vertex(graph, vertex):
    return list(graph.edges(vertex))

def vertices_with_incidence_greater_than_2(graph):
    result = [vertex for vertex, incidence in graph.degree() if incidence > 2]
    return result, len(result)

def permutation_graph_with_incidence_2(graph):
    # Get all permutations of vertices with length 2
    permuted_edges = list(permutations(graph.nodes, 2))

    # Create a new graph with permuted edges
    permuted_graph = nx.Graph(permuted_edges)

    return permuted_graph

if __name__ == "__main__":
    # Create the graph
    graph = create_graph()

    # Display the live show
    live_show(graph)

    # Calculate and print the incidence matrix
    inc_matrix = incidence_matrix(graph)
    print("\nIncidence Matrix:")
    print(inc_matrix)

    # Calculate and print the incidence list
    inc_list = incidence_list(graph)
    print("\nIncidence List:")
    print(inc_list)

    # Get a vertex from the user
    vertex = int(input("\nEnter a vertex to find edges incident to it: "))

    # Find and print edges incident to the given vertex
    incident_edges = edges_incident_to_vertex(graph, vertex)
    print(f"\nEdges incident to vertex {vertex}:")
    print(incident_edges)

    # Count and list vertices with incidence > 2
    vertices_greater_than_2, count = vertices_with_incidence_greater_than_2(graph)
    print(f"\nVertices with incidence greater than 2 ({count}):")
    print(vertices_greater_than_2)

    # Generate a permutation graph with incidence 2
    permuted_graph = permutation_graph_with_incidence_2(graph)
    print("\nPermutation Graph with Incidence 2:")
    pos_permuted = nx.spring_layout(permuted_graph)
    nx.draw(permuted_graph, pos_permuted, with_labels=True, font_weight='bold')
    plt.show()
```

```python
# e) Return the intersection graph as shown in the following figure.

import networkx as nx
import matplotlib.pyplot as plt

def create_weighted_graph_1():
    G = nx.Graph()
    G.add_edge('A', 'B', weight=7)
    G.add_edge('B', 'D', weight=4)
    G.add_edge('D', 'C', weight=3)
    G.add_edge('C', 'A', weight=2)
    return G

def create_weighted_graph_2():
    H = nx.Graph()
    H.add_edge('B', 'D', weight=4)
    H.add_edge('B', 'F', weight=5)
    H.add_edge('D', 'F', weight=6) # Common edge with graph 1
    return H

def create_intersection_graph(graph1, graph2):
    intersection_graph = nx.intersection(graph1, graph2)
    return intersection_graph

def live_show(graph, title=''):
    pos = nx.spring_layout(graph)
    labels = nx.get_edge_attributes(graph, 'weight')
    nx.draw(graph, pos, with_labels=True, font_weight='bold')
    nx.draw_networkx_edge_labels(graph, pos, edge_labels=labels)
    plt.title(title)
    plt.show()

if __name__ == "__main__":
    # Create the weighted graphs
    graph1 = create_weighted_graph_1()
    graph2 = create_weighted_graph_2()

    # Display the weighted graphs
    live_show(graph1, 'Weighted Graph 1')
    live_show(graph2, 'Weighted Graph 2')

    # Create and display the intersection graph
    intersection_graph = create_intersection_graph(graph1, graph2)
    live_show(intersection_graph, 'Intersection Graph')
```

```python
# f) Return the number of clique graphs (A clique graph is a subset of vertices of an undirected graph such that every two distinct vertices in the clique are adjacent, Application: To derive groups of people who all know each other.)

import networkx as nx

def create_weighted_graph_1():
    G = nx.Graph()
    G.add_edges_from([(1, 2), (2, 3), (1, 3), (3, 4), (4, 5)])
    return G


def live_show(graph, title=''):
    pos = nx.spring_layout(graph)
    labels = nx.get_edge_attributes(graph, 'weight')
    nx.draw(graph, pos, with_labels=True, font_weight='bold')
    nx.draw_networkx_edge_labels(graph, pos, edge_labels=labels)
    plt.title(title)
    plt.show()
    
def count_clique_graphs(graph):
    # Find all cliques in the graph
    cliques = list(nx.find_cliques(graph))

    # Count the number of cliques
    num_cliques = len(cliques)

    return num_cliques, cliques

if __name__ == "__main__":
    # Create your graph (replace this with your graph creation logic)
    graph1 = create_weighted_graph_1()
    live_show(graph1, 'Weighted Graph 1')

    # Count the number of clique graphs
    num_cliques, cliques = count_clique_graphs(graph1)

    print(f"Number of clique graphs: {num_cliques}")
    print("Clique graphs:")
    for i, clique in enumerate(cliques, 1):
        print(f"Clique {i}: {clique}")
```

```python
#g) Return a bull graph with 5 nodes, if any (A bull graph is a triangle with horns)

import networkx as nx
import matplotlib.pyplot as plt

def create_bull_graph():
    G = nx.Graph()

    # Add nodes and edges for the triangle
    G.add_edges_from([(1, 2), (2, 3), (3, 1)])

    # Add nodes and edges for the "horns"
    G.add_edges_from([(1, 4), (3, 5)])

    return G

def live_show(graph):
    pos = nx.spring_layout(graph)
    nx.draw(graph, pos, with_labels=True, font_weight='bold')
    plt.show()

if __name__ == "__main__":
    # Create the bull graph
    bull_graph = create_bull_graph()

    # Display the live show
    live_show(bull_graph)
```
